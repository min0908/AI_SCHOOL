```python
import requests
import json
import pandas as pd
from tqdm import trange
import time

import folium
from bs4 import BeautifulSoup as bs
```

```python
# URL 만들기
page_no = 1

url ="https://www.hollys.co.kr/store/korea/korStore2.do"
params = f"?pageNo={page_no}&sido=&gugun=&store="
params
```

```python
# response
response = requests.get(url, params = params)
```

```python
# response.text에 bs 적용하기
pd.read_html(response.text)[0] 
```

```python
def page_scrapping(page_no):
    url = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_no}&sido=&gugun=&store="
    response = requests.get(url)
    html = bs(response.text)
    df_table = pd.read_html(response.text)[0]
    return df_table
```

```python
page_scrapping(2)
```

```python
from tqdm import trange
import time
last_page = 52 # 할리스 마지막 페이지

requests.packages.urllib3.disable_warnings(requests.packages.urllib3.exceptions.InsecureRequestWarning)

page_list = []
for page_no in trange(1,int(last_page)+1):
    result = page_scrapping(page_no)
    page_list.append(result)
    time.sleep(0.01)
```

```python
df = pd.concat(page_list)
```

```python
df = df[['지역','매장명','현황','주소']]
```

```python
df['브랜드명'] = '할리스'
```

```python
df.head()
```

```python
df.shape
```

```python
file_name = "할리스매장.csv"
df.to_csv(file_name,index = False ,  encoding="utf-8-sig")
```

# 투썸플레이스

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome("./chromedriver")
# 투썸 모바일 버전 웹사이트 접속
driver.get("https://mo.twosome.co.kr/so/storeList.do")
# 전체매장
all_button = driver.find_element(By.CSS_SELECTOR,"body > div.wrapper.sub > div > div > form > fieldset > div.search_shop_top > div > ul > li:nth-child(2) > a")
all_button.click()
```

```python
SCROLL_PAUSE_SEC = 1

# 스크롤 높이 가져옴
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 1초 대기
    time.sleep(SCROLL_PAUSE_SEC)

    # 스크롤 다운 후 스크롤 높이 다시 가져옴
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
html = driver.page_source #페이지 소스
```

```python
bs=BeautifulSoup(html,'html.parser') # 페이지 주소 정리해서 확인
```

```python
bs
```

```python
title = bs.select('#infiniteScroll > li > div > a > dl > dt > strong')
```

```python
title
```

```python
store = []
for i in range(len(title)):
    store.append(title[i].text)
```

```python
address = bs.select('dd#addr')
```

```python
address
```

```python
store_ad = []
for i in range(len(address)):
    store_ad.append(address[i].text) 
```

```python
store_ad
```

```python
df_twosome = pd.DataFrame(zip(store, store_ad))
df_twosome.columns = ['매장명','지역']
df_twosome['브랜드명'] = '투썸플레이스'
```

```python
df_twosome
```

```python
file_name = "twosome.csv"
df_twosome.to_csv(file_name,index = False ,  encoding="utf-8-sig")
```

# 엔젤리너스

```python
page_no = 1

url = f"https://partner.payco.com/merchant/search/offline/detail/list?currentPage={page_no}&perPage=15&partnerType=FRANCHISE&partnerCode=ANGELINUS&searchWord="
```

```python
def angel_scrapping(page_no):
    url = f"https://partner.payco.com/merchant/search/offline/detail/list?currentPage={page_no}&perPage=15&partnerType=FRANCHISE&partnerCode=ANGELINUS&searchWord="
    headers = {'user-agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    angel_json = response.json()
    df = pd.DataFrame(angel_json['partnerList'])
    df = df[['partnerAddress','partnerName']]
    return df
```

```python
angel_scrapping(4)
```

```python
from tqdm import trange
import time
last_page = 21 # 엔제리너스 마지막 페이지

requests.packages.urllib3.disable_warnings(requests.packages.urllib3.exceptions.InsecureRequestWarning)

page_list = []
for page_no in trange(1,int(last_page)+1):
    result = angel_scrapping(page_no)
    page_list.append(result)
    time.sleep(0.01)
```

```python
df_angel = pd.concat(page_list)
```

```python
df_angel.columns = ['지역', '매장명']
```

```python
df_angel['브랜드명'] = '엔제리너스'
```

```python
for i in range(len(df_angel)):
    df_angel['매장명'].iloc[i] = df_angel.iloc[i]['매장명'].strip('엔제리너스( )')
```

```python
df_angel.head()
```

```python
df_angel.tail()
```

```python
file_name = "angelinus.csv"
df_angel.to_csv(file_name,index = False ,  encoding="utf-8-sig")
```
